```python
# My birthday: Aug 2nd 2002
# so I have selected book number 4
# pages from 2-11 (1811-1829) for file1
# pages from 102-111 (1969-1986) for file2

from collections import defaultdict
import re
from spellchecker import SpellChecker
from multiprocessing import Pool, cpu_count

# Initializing spell checker
spellchk = SpellChecker()

# performing Map Function
def map_function(text_chunk_in_files):
    # here, i am splitting the text into words and then map every words to ( word,1)
    words = re.findall(r'\b[a-zA-Z]+\b', text_chunk_in_files.lower())  # Extract words
    wordd_list = []
    for worditr in words:
        wordd_list.append((worditr, 1))
    return wordd_list

# performing Reduce function
def reduce_function(mapped_data):
    #here, it Aggregates word counts from the mapped key-value pairs.
    word_counts = defaultdict(int)
    for worditr, countitr in mapped_data:
        word_counts[worditr] += countitr  # Aggregate counts
    return word_counts

# identifying all the non english words
def map_non_english_words(text_chunk):
    #this will map all the non eglish words in this manner : (word,1)
    words_non_eng = re.findall(r'\b[a-zA-Z]+\b', text_chunk)
    word_list_non_eng = []
    for worditr in words_non_eng:
        if worditr.lower() not in spellchk and worditr.lower() not in spellchk.known(words_non_eng):
            word_list_non_eng.append((worditr, 1))
    return word_list_non_eng

# reading the file - function
def read_txt_file_info(file_path):
    with open(file_path, 'r', encoding='utf-8') as fl:
        return fl.read()

# processing the files with the help of mapreduce
def process_file_with_mapreduce(file_path, map_function, description):
    # Here, I am processing the file with the help of MapReduce
    print(f"\n{description} ({file_path}):")

    # Read the provided text files
    text = read_txt_file_info(file_path)

    # Dynamically adjusting the chunk size based on CPU cores that are available
    num_cores_CPU = cpu_count()
    chunk_sizex = max(len(text) // num_cores_CPU, 1000)

    # performing parallel processing
    text_chunks_in_files = []
    for ii in range(0, len(text), chunk_sizex):
        text_chunks_in_files.append(text[ii:ii+chunk_sizex])

    # Parallelizing Map Phase
    with Pool() as wrdpool:
        mapped_results = wrdpool.map(map_function, text_chunks_in_files)  # Map function runs on chunks

    # Flatten the mapped results
    all_the_flattened_mapped_data = []
    for sublist_mp in mapped_results:
        for pair in sublist_mp:
            all_the_flattened_mapped_data.append(pair)

    # Reduce Phase
    reduced_results = reduce_function(all_the_flattened_mapped_data)

    # Print sorted results
    for worditr, countitr in sorted(reduced_results.items(), key=lambda x: x[1], reverse=True):
        print(worditr, countitr)

def main():
    process_file_with_mapreduce("file1.txt", map_function, "Question 1: for all the word occurrences from page 2-11 using mapreduce")
    process_file_with_mapreduce("file2.txt", map_non_english_words, "Question 2: for all the non-English word occurrences from page 102-111")

# Run
if __name__ == "__main__":
    main()
```
